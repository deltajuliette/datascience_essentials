## Step 3: Feature extraction
1.  There are two columns that pertain to how many family members are on the boat for a given person. Create a new column called `FamilyCount` which will be the sum of those two columns.
2. Reverends have a special title in their name. Create a column called `IsReverend`: 1 if they're a preacher, 0 if they're not.
3. In order to feed our training data into a classification algorithm, we need to convert our categories into 1's and 0's using `pd.get_dummies`
  - Familiarize yourself with the [`pd.get_dummies` documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_dummies.html)
  - Create 3 columns: `Embarked_C`, `Embarked_Q` and `Embarked_S`. These columns will have 1's and 0's that correspond to the `C`, `Q` and `S` values in the `Embarked` column
  - Do the same thing for `Sex`
  - BONUS: Extract the title from everyone's name and create dummy columns

```python
# Define Loc_Reverand function
def rev_wru(name):
    if "Rev." in name:
        return 1
    else:
        return 0

# Create a column called IsReverend
raw['IsReverend'] = raw['Name'].agg(rev_wru)
raw.head(3)
```

```python
# Creating a new column called FamilyCount
raw['FamilyCount'] = raw['SibSp'] + raw['Parch']
raw.head(3)
```

```python
# Creating 3 columns Embarked_C, Embarked_Q, Embarked_S
cleaned1 = pd.get_dummies(raw, columns = ['Embarked'], prefix = 'Embarked')
cleaned1.head(3)
```

```python
cleaned2 = pd.get_dummies(cleaned1, columns = ['Sex'], prefix = 'sex')
cleaned2.head(3)
```

```python
# Extract title from everyone's name
names = raw['Name'].str.split(".", expand=True)[0].str.split(",", expand=True)
cleaned2['Title'] = names[1].str.title()
cleaned2.head()
```

```python
tidied = pd.get_dummies(cleaned2, columns = ['Title'], prefix="title")
tidied.head(3)
```

```python
# Remember the latest dataset is named tidied
# What was the overall survival rate?
survival_rate = tidied['Survived'].value_counts(normalize=True)
survival_rate.index = ['Died', 'Survived']

print(survival_rate)

print(f'The overall survival rate is {round(survival_rate["Survived"]*100, 1)}%')

# The overall survival rate is 38.4%
```

```python
# Which gender fared the worst? What was their survival rate?
survival_rate_gender = tidied.groupby("Survived")['sex_male'].value_counts(normalize=True, sort=False)
survival_rate_gender.index = [('Dead, Female'), ('Dead, Male'), ('Survived, Female'), ('Survived, Male')]

survival_rate_gender
# If sex_male == 0, then by default it is female
```

```python
# Visualisations
fig, ax = plt.subplots()

sns.countplot(data=raw, x='Survived', hue='Sex')
ax.set_xticklabels(['Died', 'Survived'])
ax.set_xlabel('')
ax.set_ylabel('Count')
plt.show()
```

```python
fig, ax = plt.subplots()

sns.barplot(x=survival_rate_gender.index, y=survival_rate_gender.values*100, palette='Blues')
ax.set_title("Breakdown of survivors / dead by gender")
ax.set_ylabel('%')
plt.show()
```

```python
# What was the survival rate for each Pclass?
survival_rate_Pclass = tidied.groupby(by="Pclass")['Survived'].value_counts(normalize=True, sort=False)
survival_rate_Pclass
```

```python
Pclass_list = ['Pclass1', 'Pclass2', 'Pclass3']
survivors_indices = []
for i in range(1,6,2):
    survivors_indices.append(survival_rate_Pclass.values[i]*100)

print(Pclass_list)
print(survivors_indices)

# Visualisations
fig, ax = plt.subplots()

sns.barplot(x=Pclass_list, y=survivors_indices, palette="Blues")
plt.ylabel("%")
plt.title("Survivors (%) of each PClass")
plt.show()
```

```python
# Did any reverends survive? How many?
survival_rate_Revs = tidied.loc[tidied['IsReverend'] == 1]['Survived'].value_counts()
survival_rate_Revs
```

```python
# What is the survival rate for cabins marked ¯\(ツ)/¯
survival_rate_UnknownCabs = tidied.loc[tidied['Cabin'] == '¯\(ツ)/¯'].groupby('Cabin')['Survived'].value_counts(normalize=True)
survival_rate_UnknownCabs
```

```python
# What is the survival rate for people whose Age is empty?
survival_rate_UnknownAge = tidied.loc[tidied['Age'].isna()]['Survived'].value_counts(normalize=True)
survival_rate_UnknownAge
```

```python
# What is the survival rate for each port of embarkation?
survival_rate_Port = raw.groupby("Embarked")['Survived'].value_counts(normalize=True, sort=False)
survival_rate_Port
```

```python
port_list = ['C', 'Q', 'S']
survivors_indices = []
for i in range(1,6,2):
    survivors_indices.append(survival_rate_Port.values[i]*100)

print(port_list)
print(survivors_indices)

# Visualisations
fig, ax = plt.subplots()

sns.barplot(x=port_list, y=survivors_indices, palette="Blues")
plt.ylabel("%")
plt.title("Survivors (%) from each port of embarkation")
plt.show()
```

```python
#What is the survival rate for children (under 12) in each Pclass?
kids = tidied.loc[tidied['Age'] < 12]
survival_rate_kids = kids.groupby("Pclass")['Survived'].value_counts(normalize = True, sort = False)
survival_rate_kids
```

```python
# Did the captain of the ship survive? Is he on the list?
capt = tidied.loc[tidied['title_ Capt'] ==1][['Survived', 'Name']]
capt
```

```python
# Of all the people that died, who had the most expensive ticket? How much did it cost?
dead_people = tidied.loc[tidied['Survived'] == 0]

most_ex_ticket = dead_people['Fare'].max()
owner_most_ex_ticket = dead_people.loc[dead_people['Fare'] == most_ex_ticket][["Name", "Fare", "Survived"]]
owner_most_ex_ticket
```

```python
# Does having family on the boat help or hurt your chances of survival?
survival_rate_Fam = tidied.groupby('FamilyCount')['Survived'].value_counts(normalize=True, sort=False)
survival_rate_Fam
```
